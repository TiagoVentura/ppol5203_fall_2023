     - a new column recoding the inter1 variable with three string labels. These labels should indicate when the iter1 variable is equal to its median, below the median and above the median values.

```python
# read
import pandas as pd
acled_sa = pd.read_csv("south_america.csv")
acled_sa.columns
```

```python
acled_sa.filter(like="event").columns.to_list()
```

```python
# 2
colnames = acled_sa.filter(like="event").columns.to_list() + acled_sa.filter(like="actor").columns.to_list()
acled_sa[colnames]

# another solution
acled_sa.filter(regex="event|actor").head(5)
```

```python
# rename
acled_sa.rename(columns={col: col.replace("event", "event_acled_south_america") for col in acled_sa.columns}).head(5)
```

```python
# 4
(acled_sa.
     assign(fatality_bin=np.where(acled_sa["fatalities"]>0, 1, 0), 
            full_location=acled_sa["region"]+ "-" + acled_sa["country"], 
            inter_recoded=np.where(acled_sa["inter1"]==acled_sa["inter1"].median(), "Median Value", 
                                  np.where(acled_sa["inter1"]>acled_sa["inter1"].median(), 
                                          "Higher Median", "Below Median"))
           ).head(5)
)
```

```python
# solutions using np.select

median_value = acled_sa["inter1"].median()
# Define the conditions
conditions = [
    acled_sa['inter1'] < median_value,
    acled_sa['inter1'] == median_value,
    acled_sa['inter1'] > median_value
]

# Define the  output values
choices = ['below median', 'at median', 'above median']

# Use np.select 
acled_sa['inter_recoded'] = np.select(conditions, choices, default='N/A')

# see
acled_sa.filter(["inter1", "inter_recoded"])
```

## Practice

Using the World Cup Matches Datasets from the Data Wrangling Notebook:
    
1. Select the following columns: MatchID, Year, Stage, Home Team Name, Away Team Name, Home Team Goals, Away Team Goals.
    
2. Clean the column names.    
    
3. Convert this data to the long format, having both a single collumn for the team playing and the goals scored  

4. Which team have played more games in the history of the world cup?

5. Which team have a higher average of goals in the history of the world cup?
    

```python
# load worldcup dataset
wc = pd.read_csv("WorldCups.csv")
wc_matches = pd.read_csv("WorldCupMatches.csv")

wc_matches

# 1 and 2
wc_copy = (wc_matches.filter(["MatchID", "Year", "Stage", "Home Team Name", "Away Team Name", "Home Team Goals", "Away Team Goals"]).
     rename(columns={col: col.lower().replace(" ", "_") for col in wc_matches.columns}).copy()
)

# 3
wc_home = (wc_copy.
    filter(["matchid", "year", "stage", "home_team_name", "away_team_name"]).
    melt(id_vars=["matchid", "year", "stage"], 
             var_name="team_host", 
            value_name="team_name").
    assign(team_host= lambda x: x["team_host"].str.replace("_team_name", ""))
)

# 3
wc_goals = (wc_copy.
    filter(["matchid", "year", "stage", "home_team_goals", "away_team_goals"]).
    melt(id_vars=["matchid","year", "stage"], 
             var_name="team_host", 
            value_name="team_goals").
    filter(["matchid","team_host", "team_goals"])
    
)

# 3
wc_tidy = pd.concat([wc_home, wc_goals],axis=1)

wc_tidy

# 4
(wc_tidy
     .groupby(["team_name"], as_index=False)
     ["team_name"]
     .size()
     .sort_values("size", ascending=False)
)        


# 5
(wc_tidy
     .groupby(["team_name"], as_index=False)
     ["team_goals"]
     .agg(["mean", "count"])
     .sort_values("mean", ascending=False). 
    head(10)
)
```
